```python
def strCleaner(string):
    string = string.lstrip()
    return string.replace('\n','')

def lineEmpty(line):
    return line == '\n'

def descripcioLlarga(line):
    if(len(line)>13):
        for space in line[0:13]:
            if(space != " "):
                return False
        else:
            return True
    else:
        return False
```

```python
with open("Factures per processar/asdf2","r") as file:
    
    lines = file.readlines()
    
    
    #Guardar l'infromació del fitxer en un diccionari
    
    factura = {"nomClient":strCleaner(lines[9]),
              "address":strCleaner(lines[10]),
              "numFactura":lines[11].split()[0],
               "data":lines[11].split()[1],
               "numClient":lines[11].split()[2],
               "codiPostal":lines[11].split()[3],
               "ciutat":lines[11].split()[4],
               "provincia":strCleaner(lines[12]),
               "nif":strCleaner(lines[14].split(":")[1]),
              }
    
    
    #Cos de la factura
    albarans = []
        
    i = 21 #Línea d'inici d'aquesta secció
    
    
    while(True):
        
        albara = {"num": lines[i].split()[2],
                  "data":lines[i].split()[4]
                 }
        
        articles = []
        
        #Següent línea
        i += 1
        
        while(True):

            article = {}
            
            if(len(lines[i])>24):
                #Detecció comença un altre albara
                
                if(lines[i][14:25] == "Albarà Núm."):
                    print(i, "Nou Albarà")
                    
                    albara["articles"] = articles
                    albarans.append(albara)
                    print(albarans)
                    break
            
            if(lineEmpty(lines[i]) and lineEmpty(lines[i+1])):
                #Detecció de final de secció
                #Si la línea a mistar està buida, i la següent també ja s'ha acabat aquesta secció
                albara["articles"] = articles
                albarans.append(albara)
                print(i, "Final secció")
                
                break
            
            else:
                #Article normal i corrent
                print(i, "Article normal")
                
                article["referencia"] = strCleaner(lines[i][0:11])
                article["descripcio"] = strCleaner(lines[i][12:50])
                article["quantitat"] = strCleaner(lines[i][51:57])
                article["preu"] = strCleaner(lines[i][58:67])
                article["import"] = strCleaner(lines[i][68:])
                
                if(descripcioLlarga(lines[i+1])):
                    #Detecció d'article llarg (descripció ocupa 2 línees)


                    article["descripcio"] += " " + strCleaner(lines[i+1])

                    i+=2
                    continue
                                                   
            
            articles.append(article)
            i+=1
            
        #Afegir albara al cos
        albarans.append(albara)
        
        #Detecció de final de secció
        #Si la línea a mistar està buida, i la següent també ja s'ha acabat aquesta secció
        if(lineEmpty(lines[i]) and lineEmpty(lines[i+1])):
            break
        
        
    factura["albarans"] = albarans
```
